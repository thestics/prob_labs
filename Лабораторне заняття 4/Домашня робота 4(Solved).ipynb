## 1 Від'ємний біноміальний розподіл

Більш детальна інформація в [документації](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.nbinom.html).

```python
from scipy.stats import nbinom
```

```python
n = 5
p = 0.4
```

### 1.1 Щільність

```python
x = np.arange(-0.5, n*5+1, 0.5)
```

Виведіть значення функції щільності від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
y = nbinom.pmf(x, n, p)
```

Зобразіть графічно значення функції щільності від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
plt.stem(x, y)
plt.show()
```

Зобразіть графічно значення функції щільності від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x1$.

```python
x1 = np.arange(0, n*5+1, 0.495)
```

```python
y1 = nbinom.pmf(x, n, p)
plt.stem(x1, y1)
plt.show()
```

### 1.2 Функція розподілу

Виведіть значення функції розподілу від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
cdfVals = nbinom.cdf(x, n, p)
print(cdfVals)

```

Зобразіть графічно значення функції розподілу від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
plt.plot(x, cdfVals, drawstyle='steps-post')
plt.show()
```

### 1.3 Випадкові величини

Згенеруйте $100$ випадкових величин з від'ємного біноміального розподілу з параметрами $n, p$.

```python
randVars = nbinom.rvs(n, p, size=100)
randVars
```

Виведіть кількість кожного елемента в отриманому випадковому векторі.

```python
from collections import Counter

Counter(randVals)
```

Зобразіть графічно кількість кожного елемента в отриманому випадковому векторі.

```python
plt.hist(randVars)
plt.show()
```

## 3 Різні параметри

### 3.1 Біноміальний розподіл

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин біноміального розподілу з параметрами:
1. $n = 15, p = 0.8$
2. $n = 15, p = 0.2$
3. $n = 20, p = 0.4$
4. $n = 20, p = 0.6$
5. $n = 7, p = 0.5$
6. $n = 25, p = 0.5$
7. $n = 100, p = 0.5$

```python
from scipy.stats import binom

nPars = (15,  15,  20,  20,  7,   25,  100)
pPars = (0.8, 0.2, 0.4, 0.6, 0.5, 0.5, 0.5)
intg = (3*i for i in range(1, 8))

plt.subplot(7, 3, 1)

x = np.arange(0, 2, 0.1)

f = plt.gcf()
f.set_size_inches(17, 25)
plt.subplots_adjust(hspace=0.4)


for i, n, p in zip(intg, nPars, pPars):
    x = np.arange(0, 2*n + 1)
    y1 = binom.pmf(x, n, p) # probability mass function
    y2 = binom.cdf(x, n, p) # cumulative distribution function
    randVars = binom.rvs(n, p, size=10_000)
    plt.subplot(7, 3, i - 2)
    plt.stem(x, y1)
    plt.plot()
    plt.subplot(7, 3, i - 1)
    plt.plot(x, y2, drawstyle='steps-post')
    plt.subplot(7, 3, i)
    plt.hist(randVars)
    plt.plot()


```

### 3.2 Геометричний розподіл

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин геометричного розподілу з параметрами:
1. $p = 0.1$
2. $p = 0.2$
3. $p = 0.3$
4. $p = 0.4$
5. $p = 0.5$
6. $p = 0.6$
7. $p = 0.7$
8. $p = 0.8$
9. $p = 0.9$

```python
from scipy.stats import geom

pVals = np.linspace(0.1, 0.9, 9)
x = np.arange(1, 20)
intg = (3*i for i in range(1, 10))

plt.subplot(9, 3, 1)

f = plt.gcf()
f.set_size_inches(17, 25)
plt.subplots_adjust(hspace=0.4)

for i, p in zip(intg, pVals):
    y1 = geom.pmf(x, p) # probability mass function
    y2 = geom.cdf(x, p) # cumulative distribution function
    randVars = geom.rvs(p, size=10_000)
    plt.subplot(9, 3, i - 2)
    plt.stem(x, y1)
    plt.plot()
    plt.subplot(9, 3, i - 1)
    plt.plot(x, y2, drawstyle='steps-post')
    plt.subplot(9, 3, i)
    plt.hist(randVars)
    plt.plot()
```

### 3.3 Розподіл Пуассона

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин розподілу Пуассона з параметрами:
1. $\mu = 0.1$
2. $\mu = 0.2$
3. $\mu = 0.3$
4. $\mu = 0.4$
5. $\mu = 0.5$
6. $\mu = 0.6$
7. $\mu = 0.7$
8. $\mu = 0.8$
9. $\mu = 0.9$

```python
from scipy.stats import poisson

muVals = np.linspace(0.1, 0.9, 9)
x = np.arange(1, 20)
intg = (3*i for i in range(1, 10))

plt.subplot(9, 3, 1)

f = plt.gcf()
f.set_size_inches(17, 25)
plt.subplots_adjust(hspace=0.4)

for i, mu in zip(intg, muVals):
    y1 = poisson.pmf(x, mu) # probability mass function
    y2 = poisson.cdf(x, mu) # cumulative distribution function
    randVars = poisson.rvs(mu, size=10_000)
    plt.subplot(9, 3, i - 2)
    plt.stem(x, y1)
    plt.plot()
    plt.subplot(9, 3, i - 1)
    plt.plot(x, y2, drawstyle='steps-post')
    plt.subplot(9, 3, i)
    plt.hist(randVars)
    plt.plot()
```
