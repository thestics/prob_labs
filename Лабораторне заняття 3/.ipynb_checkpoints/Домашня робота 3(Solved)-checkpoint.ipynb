## 1 Задача 1.3.12
Знайти ймовiрнiсть того, що серед трьох цифр, кожна з яких вибрана навмання, буде лише 1, 2, 3 рiзних.

Обчисліть теоретичну (повним перебором) та еміричну (симулюванням $100000$ експериментів) імовірності.

```python
import numpy as np
from itertools import product


def pract1(n):
    amt1 = 0  # no repeats
    amt2 = 0  # one repeat
    amt3 = 0  # two repeats (all same)
    for i in range(n):
        sample = tuple(np.random.choice(range(1, 10), 3))
        sampleLen = len(sample)
        setSampleLen = len(set(sample))
        if sampleLen == setSampleLen:
            amt1 += 1
        if sampleLen == setSampleLen + 1:
            amt2 += 1
        if sampleLen == setSampleLen + 2:
            amt3 += 1
    return amt1/n, amt2/n, amt3/n


def theor1():
    data = tuple(product(range(0, 11), repeat=3))
    size = len(data)
    amt1 = 0  # no repeats
    amt2 = 0  # one repeat
    amt3 = 0  # two repeats (all same)
    for sample in data:
        sampleLen = len(sample)
        setSampleLen = len(set(sample))
        if sampleLen == setSampleLen:
            amt1 += 1
        if sampleLen == setSampleLen + 1:
            amt2 += 1
        if sampleLen == setSampleLen + 2:
            amt3 += 1
    return amt1/size, amt2/size, amt3/size
    
print(theor1())
print(pract1(100000))
```

# 2 Задача 1.3.14
З послiдовностi чисел $1, 2, . . . , n$ вибирають навмання $k$ рiзних чисел.
Яка ймовiрнiсть того, що:
1. кожне з вибраних чисел кратне даному числу $p$;
2. кожне з вибраних чисел кратне хоча б одному з двох взаємно простих чисел $p$ i $q$;
3. серед вибраних чисел є хоча б одне кратне $p$?

Напишіть  відповідні функції для обрахунку теоретичної (повним перебором) та еміричної (симулюванням $100000$ експериментів) імовірностей в залежності від параметрів $n, k, p, q$.
Виведіть результат для
- $n = 25, k = 5, p = 3, q = 4$;
- $n = 25, k = 10, p = 3, q = 4$.

```python
from itertools import combinations


def pract2(n, k, p, q, amt):
    amt1 = 0 # divisible by p
    amt2 = 0 # divisible by p and q
    amt3 = 0 # at least one divisible by p
    for i in range(amt):
        s = np.random.choice(range(1, n + 1), k, replace=False)
        if all([n % p == 0 for n in s]): # if all divisible by p
            amt1 += 1
        if all([n % p == 0 or n % q == 0 for n in s]): # if all divisible by p or by q
            amt2 += 1
        if any([n % p == 0 for p in s]): # if any divisible by p
            amt3 += 1
    return amt1/amt, amt2/amt, amt3/amt



def theor2(n, k, p, q):
    sample = tuple(combinations(range(1, n + 1), k))
    amt1 = 0 # divisible by p
    amt2 = 0 # divisible by p and q
    amt3 = 0 # at least one divisible by p
    for s in sample:
        if all([n % p == 0 for n in s]): # if all divisible by p
            amt1 += 1
        if all([n % p == 0 or n % q == 0 for n in s]): # if all divisible by p or by q
            amt2 += 1
        if any([n % p == 0 for p in s]): # if any divisible by p
            amt3 += 1
    size = len(sample)
    return amt1/size, amt2/size, amt3/size
    

nVals = (25, 25)
kVals = (5, 10)
pVals = (3, 3)
qVals = (4, 4)
for i in range(2):
    print(f"n: {nVals[i]}, k: {kVals[i]}, p: {pVals[i]}, q: {qVals[i]}")
    s = "{}. All divisible by p: {}, all divisible by q: {}, any divisible by p:{}"
    a1, a2, a3 = pract2(nVals[i], kVals[i], pVals[i], qVals[i], 1_000)
    print(s.format("Empyric", a1, a2, a3))
    b1, b2, b3 = theor2(nVals[i], kVals[i], pVals[i], qVals[i])
    print(s.format("Theoretic", b1, b2, b3))

```

# 3 Задача 1.4.6
В коморi знаходяться $n$ пар черевикiв.
З них випадковим чином вибираються $2k$ черевикiв.
Яка ймовiрнiсть того, що серед вибраних черевикiв:
1. вiдсутнi парнi;
2. є рiвно одна комплектна пара;
3. є двi комплектнi пари?

Напишіть  відповідні функції для обрахунку теоретичної (повним перебором) та еміричної (симулюванням $100000$ експериментів) імовірностей в залежності від параметрів $n, k$.
Виведіть результат для
- $n = 8, k = 4$;
- $n = 13, k = 5$.

```python
from itertools import product, combinations


def _testNoPairs(sample):
    used = set()
    for s in sample:
        if "left" in s:
            revType = "right({})"
        else:
            revType = "left({})"
        if "left" in s:
            num = s[5:-1]
        else:
            num = s[6:-1]
        pair = revType.format(num)
        if pair not in used:
            used = used.union({s})
        else:
            return False
    return True

def _testOnePair(sample):
    used = set()
    pairs = 0
    for s in sample:
        if "left" in s:
            revType = "right({})"
        else:
            revType = "left({})"
        if "left" in s:
            num = s[5:-1]
        else:
            num = s[6:-1]
        p = revType.format(num)
        if pairs > 1:
            return False
        if p in used:
            pairs += 1
        used = used.union({s})
    return True

def _testMoreThanTwoPairs(sample):
    used = set()
    pairs = 0
    for s in sample:
        if "left" in s:
            revType = "right({})"
        else:
            revType = "left({})"
        if "left" in s:
            num = s[5:-1]
        else:
            num = s[6:-1]
        p = revType.format(num)
        if pairs >= 2:
            return True
        if p in used:
            pairs += 1
        used = used.union({s})
    return False


def theor3(n, k):
    pairs = [f"left({i})" for i in range(n)] + [f"right({i})" for i in range(n)]
    possibs = combinations(pairs, 2*k)
    size = 0
    amt1 = 0 # no pairs
    amt2 = 0 # only one pair
    amt3 = 0 # not less than two pairs
    for sample in possibs:
        size += 1
        if _testNoPairs(sample):
            amt1 += 1
        if _testOnePair(sample):
            amt2 += 1
        if _testMoreThanTwoPairs(sample):
            amt3 += 1
    return amt1/size, amt2/size, amt3/size


def empyric3(n, k):
    pairs = [f"left({i})" for i in range(n)] + [f"right({i})" for i in range(n)]
    pairs = np.random.choice(pairs, 2*n, replace=False)
    amt1 = 0 # no pairs
    amt2 = 0 # only one pair
    amt3 = 0 # not less than two pairs
    for i in range(100):
#         print(amt1, amt2, amt3)
        sample = np.random.choice(pairs, 2*k, replace=False)
#         print(sample)
        if _testNoPairs(sample):
            amt1 += 1
        if _testOnePair(sample):
            amt2 += 1
        if _testMoreThanTwoPairs(sample):
            amt3 += 1
    return amt1/100, amt2/100, amt3/100

        
for n, k in ((8, 4), (13, 5)):
    p1, p2, p3 = empyric3(n, k)
    print(f"n: {n}, k:{k}")
    print(f"Empyrical. p1:{p1}, p2:{p2}, p3:{p3}")
    p1, p2, p3 = theor3(n, k)
    print(f"Theoretical. p1:{p1}, p2:{p2}, p3:{p3}")
```

# 4 Задача 1.2.15
Нехай $\Omega = {1, 2, \ldots, 2n}$.
Всiм числам приписанi ймовiрностi, пропорцiйнi логарифмам цих чисел.
Знайти цi ймовiрностi.
Знайти ймовiрнiсть того, що в результатi експерименту з’явиться:
1. парне число;
2. непарне число.

Напишіть функцію для обрахунку еміричної (симулюванням $100000$ експериментів) імовірності в залежності від параметра $n$.
Виведіть результат для
- $n = 10$;
- $n = 25$.

```python
def theor4(n):
    probs = [np.log(x) for x in range(1, 2*n + 1)] 
    factor = 1 / sum(probs)
    probs = factor * np.array(probs)
    probOdd = sum([probs[i] for i in range(1, 2*n, 2)])
    probEven = sum([probs[i] for i in range(0, 2*n, 2)])
    return probEven, probOdd


def empyric4(n):
    probs = [np.log(x) for x in range(1, 2*n + 1)]
    factor = 1 / sum(probs)
    probs = factor * np.array(probs)
    amtEven = 0
    amtOdd = 0
    for i in range(100_000):
        num = np.random.choice(range(1, 2*n + 1), 1, p=probs)
        if num % 2 == 0:
            amtEven += 1
        else:
            amtOdd += 1
    return amtEven/100_000, amtOdd/100_000
    

nVals = (10, 25)
for n in nVals:
    pEv1, pOdd1 = theor4(n)
    print(f"theoretical\nn: {n}, even probability: {pEv1}, odd probability: {pOdd1}")
    pEv2, pOdd2 = empyric4(n)
    print(f"empyrical\nn:{n}, even probability: {pEv2}, odd probability: {pOdd2}")
```
